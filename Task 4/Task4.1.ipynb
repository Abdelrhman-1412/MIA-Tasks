```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# inserting data frame
w = pd.read_csv("Weather Dataset.csv")
w
```

```python
# correcting name
w.rename(columns={'Loud Cover': 'Cloud Cover'}, inplace= True)
w
```

```python
# cleaning data
# cleaning duplicates
w = w.drop_duplicates()
w = w[w['Temperature (C)']> -30] #removing the temp. less than -30
w = w[w['Temperature (C)']<30] #removing the temp. more than 30
w = w[w["Apparent Temperature (C)"]> -50] #removing the  apparent temp. less than -50
w = w[w["Apparent Temperature (C)"]<50] #removing the  apparent temp. more than 50
w = w[w["Pressure (millibars)"]>1000] #removing the Pressure less than 1000
w = w[w["Wind Bearing (degrees)"] >20] #removing the wind bearing less than 20
w = w[w["Cloud Cover"] == 0] #removing cloud clover not equal 0
w = w[w["Visibility (km)"] >= 0] #removing the visibility  less than 0
w = w[w["Humidity"] >= 0] #removing the HUmidity less than 0
w.sort_values(by=['Formatted Date'],ascending=True) # arranging the date 
w
```

```python
# making plot between date and temperature
plt.plot( w['Formatted Date'] ,w["Temperature (C)"])
# making date appear every 100 date on the X axis
plt.xticks(w["Formatted Date"][::100],  rotation='vertical')
# labeling the axis
plt.xlabel('time')
plt.ylabel('Temperature (C)')

# naming the plot
plt.title('plot of Temperature-Time.')
#showing the plot
plt.show()
```

```python
# making histogram of temperature
plt.hist(w["Temperature (C)"])
```

```python
# making different colors (need the size of array be the same number of elements of temperature or humidity)
colors = np.random.randint(0,200,size=len(np.array(w["Temperature (C)"])))

# making scatter plot
plt.scatter(w["Humidity"],w["Temperature (C)"], c=colors)

# labeling the axis
plt.xlabel('Humidity')
plt.ylabel('Temperature')

# give title and size for the plot
plt.title('Scatter plot of Temperature and Humidity')
plt.figure(figsize=(30, 10))

```

```python
# Drop non numerical data

w = w.drop(columns=['Formatted Date','Summary','Precip Type','Daily Summary' ,'Cloud Cover'])

# compute correlation matrix (work on numerical data only)

corr_matrix = w.corr()

# Set up the matplotlib figure

plt.figure(figsize=(8, 6))

# create heatmap

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)

# Add a title
plt.title('Correlation Heatmap')

# Show the plot
plt.show()
```

the heat map correlation indicates the existence of relationship between the Every Numerical Variable in Data and type of this relation. the numbers called Pearson correlation coefficient. which ranges from -1 (perfect negative correlation) to 1 (perfect positive correlation), with 0 indicating no linear relationship. and the darker colors indicates the strong relation while the lighter colors indicates the weak relations. as example from the heatmap : Temperature and the pressure has inverse relationship which means when that pressure increase when temperature decrease. and that barely there's no relationship between wind bearing and humidity or temperature.
